```python
df_new = df.copy()

def handleOutliers(col):
  print(col)

  print("Mean: ",df_new[col].mean())
  print("Median: ",df_new[col].median())
  print("Standard deviation: ", df_new[col].std())
  print("Min: ",df_new[col].min())
  print("Max: ",df_new[col].max())

  upper = df_new[col].mean() + 3 * df_new[col].std()
  lower = df_new[col].mean() - 3 * df_new[col].std()

  print("Upper bound: ",upper)
  print("Lower bound: ",lower)

  df_new[col] = np.where(df_new[col] > upper, upper, df_new[col])

  return df_new[(df_new[col] > upper) | (df_new[col] < lower)]
```

```python
for i in range(len(outliers)):
  handleOutliers(outliers[i])
```

```python
def plots(df, var, transformed_var):

    plt.figure(figsize=(13, 5))

    # Plot the original variable's distribution
    plt.subplot(121)
    sns.histplot(df[var], kde=True, color='blue', edgecolor='black')  # Changed to sns.histplot for modern syntax
    plt.title(f'Before Transformation: {var}', fontsize=14)

    # Plot the transformed variable's distribution
    plt.subplot(122)
    sns.histplot(transformed_var, kde=True, color='green', edgecolor='black')  # Ensure consistency
    plt.title(f'After Transformation: {var}', fontsize=14)

    # Display the plots
    plt.tight_layout()
    plt.show()
```

```python
def plots(df, var, t):

  plt.figure(figsize=(13, 5))

  plt.subplot(121)
  sns.kdeplot(df[var])
  plt.title('before' + str(t))

  p1 = t.fit_transform(df[[var]]).flatten()

  plt.subplot(122)
  sns.kdeplot(p1)
  plt.title('after' + str(t))

  plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.utils import resample
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn import metrics
from sklearn.model_selection import  train_test_split
```

```python
df = pd.read_csv('Transformed_Data.csv')
```

```python
data = df.copy()
X = data.drop('stroke', axis = 1)
Y = data.iloc[: , -1]
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size = 0.2 , random_state = 42)
```

```python
cat = [c for c in X_train if X_train[c].dtype == 'object']
for c in cat:
  le = LabelEncoder()
  le.fit(list(X_train[c]) + list(X_test[c]))
  X_train[c] = le.fit_transform(X_train[c])
  X_test[c] = le.transform(X_test[c])
```

```python
# Calculate stroke and no stroke counts
t = data[data['stroke'] == 1].shape[0]
f = data[data['stroke'] == 0].shape[0]
print(f"Stroke: {t} , No Stroke: {f}")

# Values and labels for visualization
values = [t, f]
labels = ['Stroke', 'No Stroke']

# Plotting with Seaborn
sns.barplot(x=labels, y=values, palette='coolwarm')

# Adding a title and labels
plt.title("Stroke vs No Stroke Count", fontsize=16, fontweight='bold')
plt.xlabel("Category", fontsize=12)
plt.ylabel("Count", fontsize=12)

# Show the plot
plt.show()
```

```python
def info_classification(name, y_test, y_pred, y_pred_proba=None):
    if y_pred_proba is not None:
        # Calculate ROC Curve and AUC using probabilities
        fpr_log, tpr_log, thresholds_log = metrics.roc_curve(y_test, y_pred_proba)
        auc_log = metrics.roc_auc_score(y_test, y_pred_proba)
    else:
        # Calculate ROC Curve and AUC using predictions
        fpr_log, tpr_log, thresholds_log = metrics.roc_curve(y_test, y_pred)
        auc_log = metrics.roc_auc_score(y_test, y_pred)

    # Calculate additional metrics
    f1 = metrics.f1_score(y_test, y_pred)
    f2 = metrics.fbeta_score(y_test, y_pred, beta=2)
    recall = metrics.recall_score(y_test, y_pred)
    precision = metrics.precision_score(y_test, y_pred)

    # Plotting ROC Curve
    plt.figure(figsize=(8, 6))
    plt.plot(fpr_log, tpr_log, color='m', label=f"AUC = {auc_log:.2f}")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"ROC Curve for {name}")
    plt.legend(loc=4)
    plt.grid(alpha=0.3)
    plt.show()

    # Printing Metrics
    print("\n" + name + "\n")
    print("Accuracy: " + str(metrics.accuracy_score(y_test, y_pred)))
    print("AUC: " + str(auc_log))
    print("Recall: " + str(recall))
    print("Precision: " + str(precision))
    print("F1: " + str(f1))
    print("F2: " + str(f2))
    print("Confusion Matrix: \n" + str(metrics.confusion_matrix(y_test, y_pred)))
```

```python
lg = LogisticRegression(random_state = 42)
lg.fit(X_train, Y_train)
y_pred = lg.predict(X_test)

info_classification("Simple Logistic Regression", Y_test, y_pred, None)
```
